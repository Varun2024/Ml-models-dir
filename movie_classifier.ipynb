```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\ASUS\Documents\ml pojects\Movie_classification.csv")
```

# Missing value imputation

```python
df['Time_taken'].mean()
```

```python
df['Time_taken'].fillna(value = df['Time_taken'].mean(), inplace=True)
```

```python
df.info()
```

# Dummy varible creation 

```python
df.head()
```

```python
df = pd.get_dummies(df, columns = ["3D_available", "Genre"], drop_first = True , dtype = int)
```

```python
df.head()
```

# X - y split

```python
X = df.loc[:, df.columns != "Start_Tech_Oscar"]
type(X)
```

```python
X.head()
```

```python
X.shape
```

```python
y = df["Start_Tech_Oscar"]
```

```python
y.head()
```

```python
y.shape
```

# test-train split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size = 0.2)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
X_test.shape
```

# training classification

```python
from sklearn import tree
clftree = tree.DecisionTreeClassifier(max_depth=3)
```

```python
clftree.fit(X_train,y_train)
```

# predict values using trained model

```python
y_train_pred = clftree.predict(X_train)
y_test_pred = clftree.predict(X_test)
```

# model performance

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
confusion_matrix(y_train,y_train_pred)
```

```python
confusion_matrix(y_test,y_test_pred)
```

```python
accuracy_score(y_test,y_test_pred)
```

# bagging


```python
clftree = tree.DecisionTreeClassifier()
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bag_clf = BaggingClassifier(clftree, n_estimators=1000, bootstrap =True, n_jobs= -1, random_state=42 )
```

```python
bag_clf.fit(X_train,y_train)
```

```python
confusion_matrix(y_test, bag_clf.predict(X_test))
```

```python
accuracy_score(y_test, bag_clf.predict(X_test))
```
