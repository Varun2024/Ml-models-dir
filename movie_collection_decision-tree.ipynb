## missing value imputation


```python
# we have some missing values in the time taken field so we need to impute those by using mean values(we can take mode and median values as well)
```

```python
df['Time_taken'].mean() # mean of time_taken var
```

```python
df['Time_taken'].fillna(value = df['Time_taken'].mean(),inplace = True)
```

```python
df.info()
```

 # dummy variables

```python
# for categorical varible viz, genre and 3d_available
#we need n-1 dummy var for every category in a categorical var
```

```python
df = pd.get_dummies(df,columns=['3D_available','Genre'],drop_first=True,dtype=int)  # changes in the import csv file
```

```python
df.head() # we get 1 var for 3d_available and 3 var in genre|
```

# splitting in X(independent var) and y var(taget var)

```python
# here our target var is 'Collection'
# : means all rows ,then we need to specify what col we do not need , here it is collection since it is our target var
X = df.loc[:,df.columns != "Collection"]
type(X)
```

```python
# all field are here except collection
X.head() 
```

```python
X.shape
```

```python
y = df["Collection"]
type(y)
```

```python
y.head()
```

```python
y.shape
```

# test-train split

```python
# usually we do not train our data on all values of the dataset. We leave spme portion of data for testinf our trained model
```

```python
from sklearn.model_selection import train_test_split
```

```python
# test size is 20% of data and random state a parameter that controls how data is shuffled before splitting it into training and testing sets. 
# allows reproducibilty of getting same test train split every time, always keep the random state same
X_train , X_test, y_train , y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train.head()
```

```python
# indexes are shuffled here due to splitting into test-train and 80% data in train
```

```python
X_train.shape
```

# training regression tree

```python
# well 1st create our classifier or regression object
# then well fit our x-train and y-train data into our object
# then well predict our dependent varible using our model and x-train and x-test data

#regtree object
from sklearn import tree
regtree = tree.DecisionTreeRegressor(max_depth = 3)
```

```python
regtree.fit(X_train,y_train)
```

```python
# prediction
y_train_pred = regtree.predict(X_train)
y_test_pred = regtree.predict(X_test)
```

```python
y_test_pred
```

# performance of the model 

```python
# r2 lies between 0 to 1 => 1 is best and 0 is worst , values above 0.8 indicates excellent model and between 0.4 to 0.8 are good models
from sklearn.metrics import mean_squared_error , r2_score
```

```python
# between test data and predicted data , used for somparing diff models on smae data , rn no such use for us
mean_squared_error(y_test , y_test_pred)
```

```python
# this doesnt give the actual performance
r2_score(y_train , y_train_pred) 
```

```python
# this gives actual performance
r2_score(y_test , y_test_pred)
```
